```python
dfm = bmb.Model("RT ~ decision+trial_num+planning_depth+goal_switch+delayed_planning+control_regressor+(delayed_planning*trial_num)+(decision|sub)+(trial_num|sub)+(planning_depth|sub)+(goal_switch|sub)+(control_regressor|sub)+(delayed_planning|sub)+((delayed_planning*trial_num)|sub)",data=df1)
model_fitted = dfm.fit(draws=1000, target_accept=0.99)
```
