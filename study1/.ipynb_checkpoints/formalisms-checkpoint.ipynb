### Parameters

$\beta_{MF} = \text{Model-free inverse temperature}$

###  Model-Based Controller

$$
Q_{MB_{t}}(s,a,g) = R(s,a,g) + \sum_{s'}[P(s,a,s') \cdot \text{max}_{a'}[Q_{MB}(s',a',g)]]
$$
$$
Q_{\text{Meta-Control}}(s) = 
\begin{cases}
\text{Cache } Q_{MB_{t}}(s,a,g)^* & \text{if } \text{Highest-Valued Action} = 2 \\
0 & \text{if } \text{Highest-Valued Action} < 2
\end{cases}
$$
        
$\,\,\,\,\,\,\,\,\,\, \text{if goal counter is at least 2}$

$\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\, \text{If } Q_{\text{Meta-Control}}(s) = 0$:

$$\text{Engage Planning:}$$
    
$$
    Q_{MB_{t}}(s,a,g) = R(s,a,g) + \sum_{s'}[P(s,a,s') \cdot \text{max}_{a'}[Q_{MB}(s',a',g)]]
    $$
$\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\, \text{If } Q_{\text{Meta-Control}}(s) > 0$:

$$\text{Give up control:}$$

$$ Q_{MB_{t}} = \text{Cached } Q_{MB_{t}}(s,a,g)^* $$

### Model-Free Controller
$$
Q_{MF_{t}}(s,a,g) = Q_{MF_{t-1}}(s,a,g)+\alpha(R(s,a,g)+Q_{MF_{t-1}}(s',a',g)-Q_{MF_{t-1}}(s,a,g)) 
$$

### Weighted combination of controllers

$$Q_{\text{Combined}_{t}}(s,a,g) = \beta_{MF}Q_{MF_{t}}(s,a,g)+\beta_{MB}Q_{MB_{t}}(s,a,g)$$

### Action Selection

$$ \text{Action} \sim \frac{e^{Q_{\text{Combined}_{t}}(s,a,g)}}{\sum_{a}{e^{Q_{\text{Combined}_{t}}(s,a,g)}}}$$

### Forget actions not taken

$$Q_{MF_{t}}(s,a_{\text{not taken}},g) = \gamma Q_{MF_{t-1}}(s,a_{\text{not taken},g})$$ 

### Forget actions for goals not pursued

$$Q_{MF_{t}}(s,a,g_{\text{not instructed}}) = \gamma Q_{MF_{t-1}}(s,a,g_{\text{not instructed}})$$
