
### Parameters
$\alpha = \text{learning rate}$

$\beta_{MB} = \text{Model-based weight}$

$\beta_{MF} = \text{Model-free weight}$

$\beta_{CB} = \text{Choice-bias weight}$

$\omega = \text{p(MB Failure)}$

$a=\text{choice}$

$\tilde{a}=\text{control choice}$

$g=\text{goal state}$

$s=\text{state}$

$d=\text{decision step}$


<br>
<br>



$Q_{MB}= R(s,a,g) +  \sum_{s'}[P(s,a,s') \, \text{max}_{a'}[Q_{MB}(s',a',g)_{t}]]$


$Q_{\text{MBMC}}(s,\tilde{a}=\text{relinquish control},g) = $

$0.5[Q_{MB}R(s,a=1,g)] + 0.5[Q_{MB}R(s,a=2,g)]+ R(s,\tilde{a}=\text{relinquish control},g)$

$Q_{\text{MBMC}}(s,\tilde{a}=\text{take control},g) = \text{max}_{a}[Q_{MB}R(s,a,g)]$




$\text{p(MBMC Success)}_{t}=1-\omega^{t}$

$\text{p(MBMC Failure)}_{t}=\omega^{t}$



$
\begin{flalign}
Q_{\text{MBMC-FAILURE}_{t}} = \beta_{MFMC}Q_{MFMC_{t}} +\beta_{CB}Q_{CB_{t}}
\end{flalign}
$

$
\begin{flalign}
Q_{\text{MBMC-SUCCESS}_{t}}= \beta_{MBMC}Q_{MBMC{t}}+ \beta_{MFMC}Q_{MFMC_{t}}+\beta_{CB}Q_{CB_{t}}
\end{flalign}
$

$
\begin{flalign}
LL(Q_{\text{MBMC-SUCCESS}_{t}})= \sum_{t}{[Q_{\text{MBMC-SUCCESS}_{t}}(\tilde{a}_{\text{chosen}})} - \\
ln(\sum_{i}{e^{(Q_{\text{MBMC-SUCCESS}_{t}}(\tilde{a_{i}})})}]
\end{flalign}
$

$
\begin{flalign}
LL(Q_{\text{MBMC-FAILURE}_{t}})=\sum_{t}{[Q_{\text{MBMC-FAILURE}_{t}}(\tilde{a}_{\text{chosen}})} \\ 
-ln(\sum_{i}{e^{(Q_{\text{MBMC-FAILURE}_{t}}(\tilde{a_{i}})})}]\\
\end{flalign}
$

$
\begin{flalign}
LL_{\text{total}_{t}}=ln[\text{p(MB Success)}_{t}\,e^{LL(Q_{\text{MBMC-SUCCESS}_{t}})}+\\
\text{p(MBMC Failure)}_{t}\,e^{LL(Q_{\text{MBMC-FAILURE}_{t}})}]
\end{flalign}
$

$
\begin{flalign}
Q_{MFMC_{t}}(s,\tilde{a},g) = Q_{MFMC_{t-1}}(s,\tilde{a},g)+ \\ 
\alpha(\sum_{d}{R(s,\tilde{a}_{d},g)_{t}}-Q_{MFMC_{t-1}}(s,\tilde{a},g)) 
\end{flalign}
$
